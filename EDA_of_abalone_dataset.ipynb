```python
df.column = ['Sex' , 'Length', ' Diameter', 'Hight', 'Whole_weight', 'Sucked_weight', 'Viscera_weight', 'Shell_weight', 'Rings']
df.head()
```

```python
df['Age'] = df['Rings'] + 1.5
df.head()
```

```python
plt.figure()
sns.histplot(df['Age'] , kde = True)
plt.title('Age Distrubution')
plt.xlabel('Age')
plt.ylabel('frequency')
plt.show()
```

```python
plt.figure()
sns.countplot(x = 'Sex', data = df)
plt.title ("Count By Sex")
plt.show()
```

```python
plt.figure()
sns.heatmap(df.select_dtypes(include=['number']).corr(), annot= True)
plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.figure()
sns.scatterplot(x = 'Length' , y='Age' , data = df)
plt.title('Length VS Age')
plt.show()
```

```python
print('Average Age:' , df['Age'].mean())
print('Maximum Age:' , df['Age'].max())
print('Maximum Age:' , df['Age'].min())
```
